Q6.鳥取砂丘に出現したグラードンを **"グラードン(トットリのすがた)"** として新しく登録しなきゃ！

※"グラードン(トットリのすがた)"の図鑑番号は2525です。

A6.

```python
df_6=df_index_name
df_2525=df_index_name[df_index_name['ポケモン名']=='グラードン']
df_2525.at['グラードン','図鑑番号']=2525
df_2525.at['グラードン','ポケモン名']='グラードン(トットリのすがた)'
df_2525.at['グラードン','タイプ１']='くさ'
df_2525.at['グラードン','タイプ２']='じめん'
df_2525.at['グラードン','通常特性１']='エアロック'
df_6.loc['グラードン(トットリのすがた)']=df_2525.loc['グラードン',:]
df_6.tail()
```
