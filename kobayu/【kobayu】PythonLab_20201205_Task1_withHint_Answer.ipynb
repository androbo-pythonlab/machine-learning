Q3.
2020年1月1日から2050年12月31日までの日付を8桁の数字で表現したyyyyMMdd(例:20201205)について、以下の条件を満たすものが何個あるか数え、列挙してください。

- 左右対称なもの
- 各桁の数字の和が[0,1,8]のいずれかであるもの

回答例：
- 左右対称なもの：　20コ　、　[21211212,…]
- 各桁の数字の和が[0,1,8]のいずれかであるもの：　50コ　、　[20200202,…]

A3.

# ヒント

```python
# MMとddについて、1桁のときには2桁にあわせてあげなきゃ
# 数値として扱うか？文字列として扱うか？
# 左右対称⇔「○文字目と×文字目が同じ」×4？
# 10000で割った和とあまりで左右分割できるか？
# int(20200202/10000),str(20200202%10000).zfill(4)
```

# 訂正

各桁の総和が0や1になる日付のパターンはあり得ませんでした！！

なので8のパターンを求める、

あるいは、1桁になるまで足したときの回答を出してみてください！

# 各桁の数字の総和が[0,1,8]

```python
symlist = []
l = []
for i in range(2020,2051):
    for j in range(1,13):
        for k in range(1,32):
            if j == 2 and k == 29:
                break
            if j == 4 and k == 31:
                break
            if j == 6 and k == 31:
                break
            if j == 9 and k == 31:
                break
            if j == 11 and k == 31:
                break
            temp = str(i)+str(j).zfill(2)+str(k).zfill(2)
            if temp[0] == temp[7] and temp[1] == temp[6] and temp[2] == temp[5] and temp[3] == temp[4]:
                symlist.append(temp)
            if int(temp[0])+int(temp[1])+int(temp[2])+int(temp[3])+int(temp[4])+int(temp[5])+int(temp[6])+int(temp[7]) in [0,1,8]:
                l.append(temp)
print("左右対称：",len(symlist),"コ,",symlist,"\n和が[0,1,8]:",len(l),"コ,",l)
```

# 1桁になるまで足し合わせる場合

```python
symlist = []
l = []
for i in range(2020,2051):
    for j in range(1,13):
        for k in range(1,32):
            if j == 2 and k == 29:
                break
            if j == 4 and k == 31:
                break
            if j == 6 and k == 31:
                break
            if j == 9 and k == 31:
                break
            if j == 11 and k == 31:
                break
            temp = str(i)+str(j).zfill(2)+str(k).zfill(2)
            if temp[0] == temp[7] and temp[1] == temp[6] and temp[2] == temp[5] and temp[3] == temp[4]:
                symlist.append(temp)
            tempsum = int(temp[0])+int(temp[1])+int(temp[2])+int(temp[3])+int(temp[4])+int(temp[5])+int(temp[6])+int(temp[7])
            while tempsum > 9:
                tempsum = sum(list(map(int, str(tempsum))))
            if tempsum in [0,1,8]:
                l.append(temp)
print("左右対称：",len(symlist),"コ,",symlist,"\n和が[0,1,8]:",len(l),"コ,",l)
```

同じ種類の硬貨どうしは区別できません。2 通りの硬貨の選び方は、ある種類の硬貨についてその硬貨を選ぶ枚数が異なるとき区別されます。

出力例

# ヒント

```python
# 500円から確定していくべきか、50円から確定していくべきか、、、
```

A4.

```python
a=int(input())
b=int(input())
c=int(input())
x=int(input())

combi=0
for i in range(a+1):
    for j in range(b+1):
        for k in range(c+1):
            if i*500+j*100+k*50 == x:
                combi+=1
print(combi)
```

Q5.[ABC 083 B - Some Sums](https://atcoder.jp/contests/abc083/tasks/abc083_b)

問題文

1
 以上 
N
 以下の整数のうち、
10
 進法での各桁の和が 
A
 以上 
B
 以下であるものの総和を求めてください。

A5.

```python
n,a,b=(int(x) for x in input().split())
ans=0
for i in range(1, n+1):
    if a <= sum([int(c) for c in str(i)]) <= b:
        ans += i
print(ans)
```
